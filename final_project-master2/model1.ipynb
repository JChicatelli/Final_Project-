```python
# Assign X (data) and y (target)

## BEGIN SOLUTION
X = primary_data[[
       'Mean Travel Time', 'Median Household Income', 'Below Poverty Level',
       'Total Number of Firms', '% Black-owned Firms', '% Asian-owned Firms',
       '% Hispanic-owned Firms', '% Women-owned Firms',
       'Manufacturers Shipments ($1000)', 'Retail Sales ($1000)',
       'Retail Sales per Capita', 'Building permits',
       'Land Area (Square Miles)', 'Population per Square Mile']]
y = primary_data[['Bernie Sanders', 'Hillary Clinton']]
print(X.shape, y.shape)
### END SOLUTION
```

```python
# Split the data into training and testing

### BEGIN SOLUTION
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
### END SOLUTION
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
#y_scaler = StandardScaler().fit(y_train)
```

```python
# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
#y_train_scaled = y_scaler.transform(X_train)
#y_test_scaled = y_scaler.transform(X_test)
```

```python
#from keras.utils import to_categorical

# One-hot encoding
#y_train_categorical = to_categorical(y_train)
#y_test_categorical = to_categorical(y_test)
```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=20, activation='relu', input_dim=14))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train,
    epochs=50,
    shuffle=True,
    verbose=2
)
```

# Deep Learning
For this network, we simply add an additional hidden layer of 6 nodes

```python
deep_model = Sequential()
deep_model.add(Dense(units=200, activation='relu', input_dim=14))
deep_model.add(Dense(units=200, activation='relu'))
deep_model.add(Dense(units=2, activation='softmax'))
```

```python
deep_model.summary()
```

```python
deep_model.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

deep_model.fit(
    X_train_scaled,
    y_train,
    epochs=500,
    shuffle=True,
    verbose=2
)
```

# Compare the models below

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = deep_model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
