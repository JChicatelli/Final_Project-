```python
from sklearn import tree
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join("nobetter.csv"))
df.head()
```

```python
df.columns
```

```python
df_clean = df.drop(columns=["No Preference", "Uncommitted", "Ben Carson", "Bernie Sanders", "Carly Fiorina", "Chris Christie", "Hillary Clinton"])
df_clean.head()
```

```python
df_cleaner = df_clean.drop(columns=["Jeb Bush", "John Kasich", "Marco Rubio", "Martin OMalley", "Mike Huckabee", 'Rand Paul', 'Rick Santorum', 'Ted Cruz'])
df_cleaner
```

```python
target = df_cleaner["Donald Trump"]
target_names = ["Votes"]
```

```python
data = df_cleaner.drop("Donald Trump", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test) 
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test) 
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
