```python
from sklearn import tree
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join("nobetter2.csv"))
df.head()
```

```python
df.columns
```

```python
df_clean = df.drop(columns=["('Votes', ' No Preference')", "('Votes', ' Uncommitted')","('Votes', 'Ted Cruz')", "('Votes', 'Ben Carson')",
                           "('Votes', 'Hillary Clinton')", "('Votes', 'Carly Fiorina')", "('Votes', 'Chris Christie')", "('Votes', 'Donald Trump')"
                            ])
df_clean.head()
```

```python
df_cleaner = df_clean.drop(columns=["('Votes', 'Jeb Bush')","('Votes', 'John Kasich')" , "('Votes', 'Marco Rubio')", "('Votes', 'Mike Huckabee')",
       "('Votes', 'Rand Paul')", "('Votes', 'Rick Santorum')" ])
df_cleaner.head()
```

```python
target = df_cleaner["('Votes', 'Bernie Sanders')"]
target_names = ["Votes"]
```

```python
data = df_cleaner.drop("('Votes', 'Bernie Sanders')", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test) 
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test) 
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
